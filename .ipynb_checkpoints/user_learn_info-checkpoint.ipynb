```python
#['a','b','c','d'],[1,2,3,4]-------->['a', 'b', 'b', 'c', 'c', 'c', 'd', 'd', 'd', 'd']
def time_expand(a,b):  
    c = []
    for i in range(0,len(a)):
        c += ([a[i]]*b[i])
    return c
```

```python
head = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/71.0.3578.98 Safari/537.36'
}
```

```python
def getsource(url):
    try:
        html = requests.get(url,headers=head,timeout=(3,7))
    except requests.exceptions.ConnectTimeout as e:
        print(e)
        return
    selector = etree.HTML(html.text)
    content = selector.xpath("//html")[0]
    user_id = int(url.replace('http://www.imooc.com/u/', '').replace('/courses',''))
    user_name = content.xpath('//h3[@class="user-name clearfix"]/span/text()')

    if user_name:
        user_name = user_name[0]
        user_sex = content.xpath('//p[@class="about-info"]/span/text()')
        if user_sex:
            user_sex = user_sex[0]
        else:
            user_sex = r'/'
        user_time = content.xpath('//div[@class="item follows"]//em/text()')[0].replace(' ', '') #学习时长
        user_exp = int(content.xpath('//div[@class="item follows"]//em/text()')[1])       #经验
        user_score = int(content.xpath('//div[@class="item follows"]//em/text()')[2])     #积分
        user_follows = int(content.xpath('//div[@class="item follows"]//em/text()')[3])   #关注
        user_fans = int(content.xpath('//div[@class="item follows"]//em/text()')[4])      #粉丝

        #抓取课程学习数据：1.无学习记录 2.不需要翻页 3.需要翻页

        last_page = content.xpath('//div[@class="page"]/a[last()]/text()')  #尾页？
        user_course_list = []

        #需要翻页
        if last_page:   
            if last_page[0] == u'尾页':
                last_page_num = int(content.xpath('//div[@class="page"]/a[last()]/@href')[0]
                        .replace('/u/', '').replace(str(user_id), '')
                        .replace('/courses?page=', ''))
                for i in range(1,last_page_num+1):
                    page_url = "https://www.imooc.com/u/"+str(user_id)+"+/courses?page="+str(i)
                    try:
                        page_html = requests.get(page_url,headers=head,timeout=(3,7))
                    except requests.exceptions.ConnectTimeout as e:
                        print(e)
                        return
                    page_selector = etree.HTML(page_html.text)
                    page_content = page_selector.xpath("//html")[0]

                    course_id_list = page_content.xpath('//h3[@class="study-hd"]/a/@href')
                    course_num = len(page_content.xpath('//li[@class="course-one"]'))
                    per_time_num = list(map(lambda x: len(x)
                                            ,(page_content.xpath('//div[@class="course-list course-list-m"]/ul[@class="clearfix"]'))
                                           ))
                    per_time = list(map(lambda x,y:x+'/'+y.replace('月','/').replace('日','')
                                             ,page_content.xpath('//span[@class="time"]/b/text()')
                                             ,page_content.xpath('//span[@class="time"]/em/text()')
                                            ))
                    last_time_list = time_expand(per_time,per_time_num)
                    
                    #time_num = len(content.xpath('//span[@class="time"]/b/text()'))

                    for i in range(0,course_num):
                        course_id = int(page_content.xpath('//h3[@class="study-hd"]/a/@href')[i].replace('/learn/',''))                
                        course_name = page_content.xpath('//h3[@class="study-hd"]/a/text()')[i]
                        process = page_content.xpath('//div[@class="study-points"]/span[@class="i-left span-common"]/text()')[i]
                        process = int(process.replace('已学','').replace('%',''))/100
                        note = int(page_content.xpath('//div[@class="catog-points"]/span[@class="i-left span-common"]//i/text()')[i])
                        code = int(page_content.xpath('//div[@class="catog-points"]/span[@class="i-mid span-common"]//i/text()')[i])
                        qna = int(page_content.xpath('//div[@class="catog-points"]/span[@class="i-right span-common"]//i/text()')[i])
                        last_time = last_time_list[i]
                        #user_course_list.append([user_id,user_name,user_sex,user_time,user_exp,user_score,user_follows,user_fans
                        #                        ,course_id,course_name,last_time,process,note,code,qna])
                        mydb = creat_MysqlDB()
                        sql = "INSERT INTO user_learn_info (user_id,user_name,user_sex,user_time,user_exp,\
                              user_score,user_follows,user_fans,course_id,course_name,last_time,\
                              process,note,code,qna) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s)"
                        val = (user_id,user_name,user_sex,user_time,user_exp,user_score,user_follows,user_fans
                                                ,course_id,course_name,last_time,process,note,code,qna)
                        mycursor = mydb.cursor()
                        mycursor.execute(sql, val)
                        mydb.commit()
                        

        else:
            course_num = len(content.xpath('//li[@class="course-one"]'))
            #无学习记录
            if course_num == 0:
                pass
            else:
            #不需要翻页
                course_id_list = content.xpath('//h3[@class="study-hd"]/a/@href')  
                per_time_num = list(map(lambda x: len(x)
                                            ,(content.xpath('//div[@class="course-list course-list-m"]/ul[@class="clearfix"]'))
                                           ))
                per_time = list(map(lambda x,y:x+'/'+y.replace('月','/').replace('日','')
                                             ,content.xpath('//span[@class="time"]/b/text()')
                                             ,content.xpath('//span[@class="time"]/em/text()')
                                            ))
                last_time_list = time_expand(per_time,per_time_num)
                for i in range(0,course_num):
                    course_id = int(content.xpath('//h3[@class="study-hd"]/a/@href')[i].replace('/learn/',''))                
                    course_name = content.xpath('//h3[@class="study-hd"]/a/text()')[i]
                    process = content.xpath('//div[@class="study-points"]/span[@class="i-left span-common"]/text()')[i]
                    process = int(process.replace('已学','').replace('%',''))/100  #转化为百分比
                    note = int(content.xpath('//div[@class="catog-points"]/span[@class="i-left span-common"]//i/text()')[i])
                    code = int(content.xpath('//div[@class="catog-points"]/span[@class="i-mid span-common"]//i/text()')[i])
                    qna = int(content.xpath('//div[@class="catog-points"]/span[@class="i-right span-common"]//i/text()')[i])
                    last_time = last_time_list[i]

                    #user_course_list.append([user_id,user_name,user_sex,user_time,user_exp,user_score,user_follows,user_fans
                    #                            ,course_id,course_name,last_time,process,note,code,qna])
                    mydb = creat_MysqlDB()
                    sql = "INSERT INTO user_learn_info (user_id,user_name,user_sex,user_time,user_exp,\
                          user_score,user_follows,user_fans,course_id,course_name,last_time,\
                          process,note,code,qna) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s,%s, %s, %s, %s, %s)"
                    val = (user_id,user_name,user_sex,user_time,user_exp,user_score,user_follows,user_fans
                                            ,course_id,course_name,last_time,process,note,code,qna)
                    mycursor = mydb.cursor()
                    mycursor.execute(sql, val)
                    mydb.commit()
                    
        #return(user_course_list)
#     if(html.status_code!=200):
#         print("{}:{}".format(html.status_code,url))
                
```

```python
urls = []
pool = ThreadPool(8) 
for i in range(6273234, 7194872):
    url = 'http://www.imooc.com/u/' + str(i) +'/courses'
    urls.append(url)
try:
    pool.map(getsource, urls)
except Exception as e:
    print (e)
    pool.close()
    pool.join()

pool.close()
pool.join()
```

```python
def spider(start):
    pool = ThreadPool(8) 
    for i in range(start,start+10000):
        url = 'http://www.imooc.com/u/' + str(i) +'/courses'
        try:
            pool.apply_async(getsource,(url,))
        except Exception as e:
            break
    pool.close()
    pool.join()
    print('XXXXXX')
```

```python
start = 10000
while True:
    spider(start)
    start+=10000
    print('随眠10s')
    #time.sleep(8)
```
